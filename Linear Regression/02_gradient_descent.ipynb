```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
X = np.arange(10)
Y = (X - 5) ** 2 + 3         # (X-5)^2 + 3 function we take
print(X)
print(Y)
```

```python
#Visualize this curve
plt.plot(X, Y)
plt.xlabel('X')
plt.ylabel('Y')
plt.show
```

We take function Y=(X-5)^2+3  
So this function is minimum when we have X=5, Y=(5-5)^2 + 3 = 3  

So minimum value is X=5,Y=3  

We will achieve this value by using Gradient Descent.  

```python
#Gradient Descent
#init x with any random value
# y = (x - 5) ** 2 + 3
# dy/dx = 2*(x-5)
# x=x-learning_rate*grad

x = 0
lr = 0.1
plt.plot(X,Y)
for i in range(50):
    grad = 2*(x-5)
    x = x - lr * grad
    y = (x - 5) ** 2 + 3
    print(f"Gradient at step {i}: {grad} and x: {x}")
    plt.scatter(x,y)

plt.show()
```

Starting from X=0 we take bigger steps but as soon as we move towards X= 5 steps are not that much big because dy/dx is less

lets starts from other end X=9 for better understanding
