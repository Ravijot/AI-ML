```python
#Create a function for Data Generation
def generate_data(m):
    X = np.random.randn(m) * 10  # Data is between -10 and 10
    noise = np.random.randn(m)
    Y = 3*X + 1 + noise*10
    return X, Y

X, Y = generate_data(100)
print(X.shape)
print(Y.shape)
```

We need to normalize the data when we have multiple featuresvbecause they measured on different scales. Here we have only one feature still we normalize it with formula using x=x-x.mean()/x.std()

```python
def normalize_data(X):
    return (X - np.mean(X)) / np.std(X)

X = normalize_data(X)
plt.scatter(X, Y)
plt.show()
```

Now we can see the center of result is change from -10 to 10 to -1 to 1
