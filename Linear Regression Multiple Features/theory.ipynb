Notation (multi-feature)

m → number of training examples (here 4)

n → number of features (can be ≥1)

X → input matrix of shape m × n (without intercept)

X̄ → input matrix with intercept (bias term) of shape m × (n+1), where the first column is all 1s

y → target vector of shape m × 1

θ → weights (including bias) of shape (n+1) × 1

ŷ = X̄ θ → hypothesis (vector of predictions)



Hypothesis Function  
hypothesis_theta = theta_0 + X1_theta_1 + X2_theta_2 + X3_theta3  




Here we are using matrix multipication not for loops because for loops will take lot of time to execute.


We define the feature vector **x** (including the dummy feature \(x_0 = 1\)) as:


\begin{bmatrix}
x_0 \\
x_1 \\
x_2 \\
x_3
\end{bmatrix}


The parameter vector **θ** is:


\begin{bmatrix}
\theta_0 \\
\theta_1 \\
\theta_2 \\
\theta_3
\end{bmatrix}


The transpose of **θ** is:


\begin{bmatrix}
\theta_0 & \theta_1 & \theta_2 & \theta_3
\end{bmatrix}


The hypothesis is the dot product:


theta*x = scalar quantity


Which expands to:


theta * x = theta_0 * x_0 + theta_1 * x_1 + theta_2 * x_2 + theta_3 * x_3
